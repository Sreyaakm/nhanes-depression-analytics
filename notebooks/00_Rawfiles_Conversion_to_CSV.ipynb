# Task
Converting all necessary .xpt files in the current directory to .csv format and provide guidelines making it ready to download the generated .csv files.

## Step 1: Identifing and listing all .xpt files in the directory


Here, using the glob module to find all .xpt files in the current directory and store them in a list.



```python
import glob
allxptfileslisted = glob.glob('*.xpt')
print(allxptfileslisted)
```

## Step 2: Converting .xpt to .csv
Here, Iterating through the list of .xpt files using for loop, read each one into a pandas DataFrame, and save it as a .csv file without including the index in the output.


```python
for each_xpt_file in allxptfileslisted:
    df = pd.read_sas(each_xpt_file)
    csv_file = each_xpt_file.replace('.xpt', '.csv')
    df.to_csv(csv_file, index=False)
```

## Step 3: Providing guidelines & Downloading .csv files

List the generated .csv files and giving instructions on how to download them.


```python
import os
#The above line imports the os module, which provides an interface to the operating system, including access to files and directories.

all_csv_files = [f for f in os.listdir('.') if f.endswith('.csv')]
print("Generated CSV files:")
for csv_file in all_csv_files:
    print(csv_file)

print("\n In order to download .csv files which are converted, use the file browser in the left sidebar of the Google Colab environment. Check and navigate to the file list and right click on the particular .csv file and select download then, your desired .csv file gets downloaded and will be ready to use. ")
```
