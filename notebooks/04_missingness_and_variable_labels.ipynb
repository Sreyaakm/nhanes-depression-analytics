### Step 4: Missingness per variable (count + proportion)

```python
missing_by_var = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_pct": df.isna().mean() * 100
}).sort_values("missing_pct", ascending=False)

missing_by_var
```

**Comments:** This table summarizes the number and percentage of missing values for each variable in the merged dataset, sorted from highest to lowest missingness. Several demographic variables (e.g., AgeMonths_RIDAGEMN, AgeAtExamMonths_RIDEXAGM, EducationYouth_DMDEDUC3) show very high missingness because they only apply to specific subgroups (e.g., infants or youth). The PHQ-9 items (DPQ010â€“DPQ100) each have about ~8% missingness, which is relatively consistent across questions. Key identifiers and core demographics (e.g., SEQN, Sex_RIAGENDR, AgeYears_RIDAGEYR, RaceEthnicity variables) have no missing values. This pattern suggests that most missingness is structural (not applicable) rather than due to data corruption, but variables with very high missingness may need to be excluded or handled carefully in later analyses.

### Step 5: Missingness per participant

```python
df["missing_per_participant"] = df.isna().sum(axis=1)
df[["SEQN", "missing_per_participant"]].head()
```

** Comments: ** This step calculates how many variables are missing for each participant in the merged dataset. The missing_per_participant column represents the total number of missing values across all variables for that individual. For example, SEQN 93706 has 13 missing values, while SEQN 93708 has 6 missing values. This helps identify participants with a large amount of missing data who may need to be excluded from certain analyses. Examining missingness at the participant level is important to ensure that analytic results are not driven by individuals with very incomplete records.

### Step 6: Checking if missingness is related to depression status

```python
df.groupby("probable_depression")["missing_per_participant"].agg(["count", "mean", "median", "max"])
```

**Comments:** (Missingness vs. Depression Status):
This table summarizes the amount of missing data per participant, grouped by depression status. Participants without probable depression (0) have an average of about 8.12 missing variables, while participants with probable depression (1) have a slightly lower average of about 6.90 missing variables. The median number of missing variables is the same for both groups (7), suggesting similar typical levels of missingness. The maximum number of missing variables is higher among non-depressed participants (29) compared to depressed participants (14). Overall, missingness does not appear to be substantially higher among depressed individuals, indicating that excluding participants with missing data is unlikely to disproportionately remove depressed participants from the analytic sample.
