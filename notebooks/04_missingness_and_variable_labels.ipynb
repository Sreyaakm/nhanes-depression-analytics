### Step 6: Missingness per participant

```python
final_df["missing_per_participant"] = final_df.isna().sum(axis=1)
final_df[["SEQN", "missing_per_participant"]].head()
```

**Comments:** This step calculates how many variables are missing for each participant in the cleaned dataset. The missing_per_participant column represents the total number of missing values across all selected variables for that individual. For example, SEQN 93706 has 7 missing values, SEQN 93709 has 11 missing values, while SEQN 93711 has 0 missing values, indicating a complete record. This helps identify participants with higher levels of incomplete data who may need to be reviewed or handled carefully in later analyses. Examining missingness at the participant level ensures that results are not overly influenced by individuals with large amounts of missing information.

### Step 7: Checking if missingness is related to depression status

```python
final_df.groupby("probable_depression")["missing_per_participant"].agg(["count", "mean", "median", "max"])
```

**Comments:** (Missingness vs. Depression Status): This table summarizes the number of missing variables per participant, grouped by depression status. Participants without probable depression (0) have an average of about 2.23 missing variables, while participants with probable depression (1) have a lower average of about 1.20 missing variables. The median number of missing variables is the same for both groups (1), indicating similar typical levels of missingness. The maximum number of missing variables is higher among non-depressed participants (16) compared to depressed participants (5). Overall, missingness appears slightly higher among non-depressed individuals, suggesting that removing participants with missing data is unlikely to disproportionately exclude those with probable depression from the analytic sample.
